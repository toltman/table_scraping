```python
def lowest_level(row):
    location = ""
    row_levels = [f"row_level_{x}" for x in range(1,8)]
    for row_level in row_levels:
        if row[row_level] != "":
            location = row[row_level]
    return location
```

```python
tab.row_info['location'] = tab.row_info.apply(lowest_level, axis=1)
```
