- loop through columns
- determine which ones contain footnotes \d\

```python
def is_fn_col(col):
    """col contains only footnotes"""
    
    return col.str.contains(r"\\[0-9]\\").any()
```

```python
fn_cols = df.apply(lambda x: is_fn_col(x), axis=0)
```

```python
for i in fn_cols.index:
    if fn_cols[i]:
        fn_col = df.loc[:, i].fillna("")
        prev_col = df.loc[:, i-1].fillna("").astype(str)
        df.loc[:, i-1] = prev_col + fn_col

df.loc[:, ~fn_cols]
```
