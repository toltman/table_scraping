```python
def get_leading_spaces(string):
    string = str(string)
    res = re.search(r"[^ ]", string)
    
    if res:
        return res.start()
    else:
        return 0
        
```

```python
def row_info(table):
    subtitle = ""
    row_level = 0
    indent_level = 0
    rows = table.end_row - table.header_lines
    
    empty = np.empty([rows, table.ROW_LEVELS])
    empty[:] = np.NaN
    
    row_levels = pd.DataFrame(empty,
                              index=range(table.header_lines+1, table.end_row+1))

    for row in range(table.header_lines + 1, table.end_row + 1):
        cell = table.sheet.cell(row, 0)
        cell_xf = table.book.xf_list[cell.xf_index]
        is_bold = bool(table.font[cell_xf.font_index].bold)
        is_empty = bool(cell.value == "")
        indents = get_leading_spaces(cell.value)
        is_total = is_bold and (indents == 3 or indents == 5)

        
        if is_empty and table.sheet.cell(row,1).value != "":
            subtitle = table.sheet.cell(row, 1).value
        
        
        if indents in [0,3,5]:
            indent_level = 0
        else:
            indent_level = indents / 2
        
        
        if is_total:
            row_level = 0
            row_levels.loc[row, "subtitle"] = subtitle
            row_levels.loc[row, row_level] = cell.value
            row_level = 1
        elif is_bold:
            row_level = 0
            row_levels.loc[row, "subtitle"] = subtitle
            row_levels.loc[row, row_level] = cell.value
            row_level = 1
        else: 
            row_levels.loc[row, "subtitle"] = subtitle
            row_levels.loc[row, row_level+indent_level] = cell.value
        
    # forward fill row levels
    row_levels = row_levels.ffill(axis=1).ffill(axis=0)
    is_duplicate = row_levels.apply(lambda row: row.duplicated(), axis=1)
    row_levels = row_levels.where(~is_duplicate, "")
    
    return row_levels
```
