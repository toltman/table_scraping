```python
import numpy as np
import pandas as pd
import re
import os
from table import Table
```

```python
pd.set_option("display.max_rows", None, "display.max_columns", None)
```

```python
tabn = '105.30'
tab = Table('100tables/tabn' + tabn + '.xls', '2019')
```

```python
tab.table_info
```

```python
tlines = tab.title_lines
cols = list(range(0,tab.sheet.ncols))
header = pd.read_excel(tab.filename,
                        skiprows=tlines,
                        header=None,
                        nrows=tab.header_lines - tlines,
                        usecols=cols
                        )
```

```python
header
```

```python
header = header.iloc[:, 1:]
header
```

```python
header = header.ffill(axis=0).ffill(axis=1)
header
```

```python
pd.MultiIndex.from_arrays(header.values)
```

```python
header = pd.MultiIndex.from_arrays(header.values)
```

```python
col_info = header.to_frame(index=False)
```

```python
col_info
```

```python
is_duplicate = col_info.apply(lambda row: row.duplicated(), axis=1)
col_info = col_info.where(~is_duplicate, "")
col_info
```

```python
col_info.shape
```

```python
for n in range(col_info.shape[1], 7):
    col_info.insert(n, n, "")
```

```python
col_info.columns = [f"column_level_{col+1}" for col in col_info.columns]
```
